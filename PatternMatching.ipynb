```python
answer = [1, 2]
match answer:
    case []:   
        print('пустой список')
    case [x]:
        print('список с одним элементом x = {0}'.format(x))
    case [x, _]:
        print('много элементов, но первый все равно x = {0}'.format(x))
```

```python
match [0, 1, 2, 3]: 
    case [*_, x, _]:
        print(x)
```

```python
def fib(n: int) -> int: 
    if n == 0: 
        return 0
    if n == 1: 
        return 1
    return fib(n - 1) + fib(n - 2)
```

```python
fib(10)
```
